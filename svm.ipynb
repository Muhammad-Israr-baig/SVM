# Importig the essentials libaries

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
from sklearn import svm
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

# Loading the dataset

```python
iris= datasets.load_iris()

X= iris.data
y= iris.target

print(X,y)
```

# Visualizationing the dataset

```python
plt.plot(X,y)
```

# Adding features

```python
classes =[ 'iris setisa' ,' iris versicolor', ' iris virgincia']

print(X.shape)
print(y.shape)
```

# Splitting the dataset in train and test

```python
x_train ,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

# Calling the model and fitting it

```python
model =svm.SVC()
model.fit(x_train, y_train)
print(model)
```

# Making predictions and checking accurarcy

```python
predictions= model.predict(X_test)
acc= accuracy_score(y_test, predictions)
print("predictios: ", predictions)
print("accuracy: ", acc)
```
